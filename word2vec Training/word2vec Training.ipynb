```python
stopWords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Helper to convert NLTK POS tags to WordNet POS
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default to noun

def wordNetLemmatizerWithoutStopwords(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())  # Lowercase and remove punctuation
    words = word_tokenize(text)
    pos_tags = pos_tag(words)

    lemmatized = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag)) if word not in stopWords else word
        for word, tag in pos_tags
    ]
    return ' '.join(lemmatized)
```

```python
dataset['question1'] = dataset['question1'].apply(wordNetLemmatizerWithoutStopwords)
dataset['question2'] = dataset['question2'].apply(wordNetLemmatizerWithoutStopwords)
```

```python
import gensim
from gensim.models import Word2Vec
import pickle
```

```python
### i don't remove stopword because of context preservation
### remove stopwords from questions dataframe
stopWord = set(stopwords.words('english'))
def preprocess(text):
  tokens = word_tokenize(text)
  # Remove stopwords
  tokens = [word for word in tokens if word not in stopWord]
  return tokens
```

```python
q1Tokens = ques_df['question1'].apply(preprocess)
q2Tokens = ques_df['question2'].apply(preprocess)
```

```python
## train word2vec model
# q1Tokens = dataset['question1']
# q2Tokens = dataset['question2']

sentences = q1Tokens.tolist() + q2Tokens.tolist()

model = Word2Vec(sentences, vector_size = 100, window = 5, min_count = 2, workers = 4)

```

```python
pickle.dump(model, open('TrainedWord2Vec4.pkl', 'wb'))
```

```python
import numpy as np

def avg_word2vec(sentence, model, vector_size = 100):
    words = preprocess(sentence)
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    if len(word_vectors) == 0:
        return np.zeros(vector_size)
    return np.mean(word_vectors, axis=0)
```

```python
# For both questions
q1_vectors = ques_df['question1'].apply(lambda x: avg_word2vec(str(x), model))
q2_vectors = ques_df['question2'].apply(lambda x: avg_word2vec(str(x), model))
```

```python
q1_vectors[0].shape
```

```python
import pandas as pd

# Convert list of vectors to DataFrame
q1DF = pd.DataFrame(q1_vectors.tolist(), columns=[f'q1W2v{i}' for i in range(100)])
q2DF = pd.DataFrame(q2_vectors.tolist(), columns=[f'q2W2v{i}' for i in range(100)])

# Combine both question vectors
questionsAfterVectorTransformation = pd.concat([q1DF, q2DF], axis=1)
```
